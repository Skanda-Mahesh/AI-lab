```python
class Fact:
    def __init__(self, name, value=True):
        self.name = name
        self.value = value

    def __eq__(self, other):
        if not isinstance(other, Fact):
            return NotImplemented
        return self.name == other.name and self.value == other.value

    def __hash__(self):
        return hash((self.name, self.value))

    def __repr__(self):
        return f"Fact({self.name}={self.value})"

class Rule:
    def __init__(self, premises, conclusion):
        self.premises = premises  # List of Fact objects
        self.conclusion = conclusion # Fact object

    def __repr__(self):
        return f"Rule({[str(p) for p in self.premises]} -> {self.conclusion})"


def forward_chaining(facts, rules):
    new_facts = set(facts)
    tmp = set()

    while True:
        added_new_fact = False
        for rule in rules:

            premises_met = all(p in new_facts for p in rule.premises)

            if premises_met and rule.conclusion not in new_facts:
                new_facts.add(rule.conclusion)
                tmp.add(rule.conclusion)
                added_new_fact = True

        if not added_new_fact:
            break
    return list(new_facts), list(tmp)



initial_facts = {Fact("dargeling_is_holiday_spot"), Fact("munar_is_highly_visited"), Fact("dargeling_is_highly_visited")}

rules = [
    Rule([Fact("is_holiday_spot")], Fact("has_ice_creams")),
    Rule([Fact("is_highly_visited")], Fact("is_popular")),
]

all_facts, inferred_facts = forward_chaining(initial_facts, rules)


ice_cream_available = Fact("has_ice_creams") in inferred_facts or rules

if ice_cream_available:
    print("Ice cream avalibale")
```
